* A heart attack (medically known as a myocardial infarction) is a deadly medical emergency where your heart muscle begins to die because it isn’t getting enough blood flow. 
* This is usually caused by a blockage in the arteries that supply blood to your heart. If blood flow isn’t restored quickly, a heart attack can cause permanent heart damage and death.

* Every year, about 805,000 people in the United States have a heart attack.
    Of these,
    - 605,000 are a first heart attack
    - 200,000 happen to people who have already had a heart attack
    - About 1 in 5 heart attacks are silent—the damage is done, but the person is not aware of it.
 
 

**[According to World Health Organization (WHO)](https://www.who.int/health-topics/cardiovascular-diseases#tab=tab_1)**

Cardiovascular diseases (CVDs) are the leading cause of death globally, taking an estimated 17.9 million lives each year. CVDs are a group of disorders of the heart and blood vessels and include coronary heart disease, cerebrovascular disease, rheumatic heart disease and other conditions. More than four out of five CVD deaths are due to heart attacks and strokes, and one third of these deaths occur prematurely in people under 70 years of age.

The most important behavioural risk factors of heart disease and stroke are unhealthy diet, physical inactivity, tobacco use and harmful use of alcohol. The effects of behavioural risk factors may show up in individuals as raised blood pressure, raised blood glucose, raised blood lipids, and overweight and obesity. These “intermediate risks factors” can be measured in primary care facilities and indicate an increased risk of heart attack, stroke, heart failure and other complications.

Often, there are no symptoms of the underlying disease of the blood vessels. A heart attack or stroke may be the first sign of underlying disease. Symptoms of a heart attack include:
* pain or discomfort in the centre of the chest; and/or
* pain or discomfort in the arms, the left shoulder, elbows, jaw, or back.
In addition the person may experience difficulty in breathing or shortness of breath; nausea or vomiting; light-headedness or faintness; a cold sweat; and turning pale. Women are more likely than men to have shortness of breath, nausea, vomiting, and back or jaw pain.

The most common symptom of a stroke is sudden weakness of the face, arm, or leg, most often on one side of the body. Other symptoms include sudden onset of:
* numbness of the face, arm, or leg, especially on one side of the body;
* confusion, difficulty speaking or understanding speech;
* difficulty seeing with one or both eyes;
* difficulty walking, dizziness and/or loss of balance or coordination;
* severe headache with no known cause; and/or
* fainting or unconsciousness.

* **Age** : Age of the patient
* **Sex** : Sex of the patient
    - Value 0: Fele
    - Value 1: Male
* **cp** : Chest Pain type
    - Value 0: typical angina
    - Value 1: atypical angina
    - Value 2: non-anginal pain
    - Value 3: asymptomatic

* **trtbps** : resting blood pressure (in mm Hg)

* **chol** : cholesterol in mg/dl fetched via BMI sensor

* **fbs** : (fasting blood sugar > 120 mg/dl)
    -  Value 1 = true
    -  Value 0 = false

* **rest_ecg** : resting electrocardiographic results
    - Value 0: normal
    - Value 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)
    - Value 2: Hypertrophy by Estes' criteria (is a disease in which the heart muscle becomes thickened.The thickened heart muscle can make it harder for the heart to pump blood.)

* **thalach** : Maximum Heart Rate Achieved

* **exang** : exercise induced angina
    -  Value 1 = yes
    -  Value 0 = no

* **old peak** : ST depression induced by exercise relative to rest

* **slp** : the slope of the peak exercise ST segment
    -  Value 0 = unsloping
    -  Value 1 = flat
    -  Value 2 = downsloping

* **caa** : number of major vessels (0-3) colored by flourosopy.

* **thal** : thalassemia
    -  Value 0 = null
    -  Value 1 = fixed defect
    -  Value 2 = normal
    -  Value 3 = reversable defect

* **output/target** : diagnosis of heart disease (angiographic disease status)
    -  Value 0:  less chance of heart disease
    -  Value 1:  more chance of heart disease

```python
# Creating a plotting function for plotting Facetgrid graphs.

def my_facetgrid_plot(numerical_features,numerix_axis_label):
    title_font = {'color':'black','size':18, 'weight':'bold'}
    axis_font = {'color':'grey','size':16, 'weight':'bold'}
    
    sns.set_style('whitegrid')

    # custom 'ylim' value so the plotiing remains b/w these values    
    g = sns.FacetGrid(df,hue = 'target',height = 6, xlim= (df[numerical_features].min() - 10,df[numerical_features].max() + 10) )
    g.map(sns.kdeplot,numerical_features, shade = True )
    g.add_legend()
    
    plt.title(numerical_features, fontdict=title_font)
    plt.xlabel(numerix_axis_label, fontdict=axis_font)
    plt.ylabel('Density',fontdict=axis_font)
    plt.show()   
    
```

`Note :`

The KDE plot shows the density of the feature for each value of the target variable. There are usually two-colored humps representing the two values of the Target variable. If the humps are well-separated and non-overlapping, there is a correlation with the Target. If the humps overlap a lot, that means the feature is not well-correlated with the Target because the Target is equally as common on those values of the feature.

```python
my_facetgrid_plot(numerical_features[0],numerix_axis_label[0])
```

### `Dealing with Outliers`
1. 1st Way : Delete the Outliers.( `Data Deletion` )
2. 2nd Way : Value Assignment.
3. 3rd Way : Apply Limitation on the top and bottom of the variable `winsorize`
4. 4th Way : Inter Quartile (IQR) method


**trtbps feature** using `winsorize method`

* When feature is very close to normal distribution but has outliers so, we use `zscore method` to find the outliers.

    `zscore = (observations - mean) / (std deviation)`

```python
# Import the required libraries
from scipy import stats
from scipy.stats import zscore
from scipy.stats.mstats import winsorize
```

```python
# Calculate threshold value
z_score_trtbps = zscore(df['trtbps'])

# Find the outliers based on threshold values
for threshold in range(1,4):
    print(f"Threshold Value : {threshold}")
    print(f"No. Of Outliers : {len(np.where(z_score_trtbps > threshold)[0])}\n")
    print('*'*30)
    
```

```python
# Outliers with threshold value > 2
df[z_score_trtbps > 2][['trtbps']]
```

```python
# Get the least min outlier
df[z_score_trtbps > 2]['trtbps'].min()
```

```python
# Get the trtbps value < 170
df[df['trtbps'] < 170]['trtbps'].max()
```

```python
# Compute the percentile rank of a score relative to a list of scores.
winsorize_percentile_trtbps = (stats.percentileofscore(df['trtbps'],165))/100
winsorize_percentile_trtbps
```

A `percentileofscore` of, for example, 96% means that 96% of the
scores in `df[trtbps]` are below the given score. In the case of gaps or
ties, the exact definition depends on the optional keyword, `kind`.

```python
# The Value we will use for the Upper Limit
upper_limit = 1 - winsorize_percentile_trtbps 
upper_limit
```

```python
# Use winsorize function that returns a Winsorized version of the input array
trtbps_winsorize = winsorize(df['trtbps'],(0,upper_limit))
```

```python
fig, ax = plt.subplots(1,2,figsize=(18,8))
ax[0].boxplot(df['trtbps'])
ax[0].set_title("'trtbps' Before dealing with Outliers",fontdict = title_font_style)
ax[0].set_xlabel("trtbps",fontdict = label_font_style)

ax[1].boxplot(trtbps_winsorize)
ax[1].set_title("'trtbps' After dealing with Outliers",fontdict = title_font_style)
ax[1].set_xlabel("trtbps",fontdict = label_font_style)

plt.show()
```

```python
# Add trtbps_winsorize into the 'df'
df['trtbps_winsorize'] = trtbps_winsorize
df.head()
```

**thalach feature** using `IQR`

```python
# Returns outliers with the help of IQR Method
def my_iqr(df,col):
    q1 = np.quantile(df[col],0.25)
    q3 = np.quantile(df[col],0.75)
    diff = q3 - q1 #IQR
    
    lower_limit = q1 - (1.5 * diff)
    upper_limit = q3 + (1.5 * diff)
    
     # Returning the outliers
    return df[(df[col] < lower_limit) | (df[col] > upper_limit)]
```

```python
thalach_outliers = my_iqr(df,'thalach')
thalach_outliers
```

#### ` AUC (Area Under The Curve) ROC (Receiver Operating Characteristics) curve`
<img src='images/roc_auc_curve.png' width = 600>

* In Machine Learning, performance measurement is an essential task. So when it comes to a classification problem, we can count on an AUC - ROC Curve. 
* It is one of the most important evaluation metrics for checking any classification model’s performance. It is also written as `AUROC (Area Under the Receiver Operating Characteristics)`


* AUC - ROC curve is a performance measurement for the classification problems at various threshold settings. ROC is a probability curve and AUC represents the degree or measure of separability. It tells how much the model is capable of distinguishing between classes. Higher the AUC, the better the model is at predicting 0 classes as 0 and 1 classes as 1. By analogy, the Higher the AUC, the better the model is at distinguishing between patients with the disease and no disease.

**Random Classifier :** A classifier with the random performance level always shows a straight line from the origin (0.0, 0.0) to the top right corner (1.0, 1.0). Two areas separated by this ROC curve indicates a simple estimation of the performance level.

* The ROC curve is plotted with TPR (True Positive Rate) against the FPR (Fallse Positive Rate) where TPR is on the y-axis and FPR is on the x-axis.

```python
# Plot the ROC-AUC curve
fig,ax = plt.subplots(figsize=(18,8))

RocCurveDisplay.from_estimator(logistic_regression,X_test,y_test,name='Logistic Regression',ax=ax )
plt.title('Logistic Regression ROC - AUC curve',fontdict=title_font_style)
plt.plot([0,1],'r--')
plt.xlabel('False Positive Rate (Positive label : 1)',fontdict=label_font_style)
plt.ylabel('True Positive Rate (Positive label : 1)',fontdict=label_font_style)

plt.show()
```
