#### Custom displot

```python
def displot(data, x) -> None:
    sns.displot(data = data, x = x, kde = True, height = 5, color= "royalblue")
    plt.title(f"{x.upper()}", fontdict=title_style)
    plt.grid(axis="y", alpha = 0.7)
    plt.xlabel(x, fontdict=label_style)
    plt.ylabel("frequency", fontdict=label_style)
    plt.axvline(data[x].mean(), color = "orange", ls = "--")
    plt.show()
```

#### Plot all the numerical features at once

```python
for i in numerical_features:
    displot(data,i)
```

### Outliers

```python
# plotting numerical features
fig,ax = plt.subplots(1,4, figsize = (18,8))
ax[0].boxplot(data['age'])
ax[0].set_title('Age'.upper(), fontdict = title_style)

ax[1].boxplot(data['trtbps'])
ax[1].set_title('trtbps'.upper(), fontdict = title_style)

ax[2].boxplot(data['thalach'])
ax[2].set_title('thalach'.upper(), fontdict = title_style)

ax[3].boxplot(data['old peak'])
ax[3].set_title('old peak'.upper(), fontdict = title_style)
plt.show()
```

#### IQR (inter quartile range): trtbps 

```python
# handling outliers
q1 = np.percentile(data.trtbps, 25)
q3 = np.percentile(data.trtbps, 75)

iqr = q3 - q1

lower_limit = q1 - (1.5 * iqr)
upper_limit = q3 + (1.5 * iqr)

data = data[(data.trtbps > lower_limit) & (data.trtbps < upper_limit)]
data
```

#### IQR (inter quartile range): thalach 

```python
# handling outliers
q1 = np.percentile(data.thalach, 25)
q3 = np.percentile(data.thalach, 75)

iqr = q3 - q1

lower_limit = q1 - (1.5 * iqr)
upper_limit = q3 + (1.5 * iqr)

data = data[(data.thalach > lower_limit) & (data.thalach < upper_limit)]
data
```

#### IQR (inter quartile range): old peak

```python
# handling outliers
q1 = np.percentile(data["old peak"], 25)
q3 = np.percentile(data["old peak"], 75)

iqr = q3 - q1

lower_limit = q1 - (1.5 * iqr)
upper_limit = q3 + (1.5 * iqr)

data = data[(data["old peak"] > lower_limit) & (data["old peak"] < upper_limit)]
data
```

### Skewness

```python
numerical_features = ["age","trtbps","thalach","old peak"]
data[numerical_features].agg(["skew"]).T
```

Old Peak is positively or right skewed, now lets confirm visually.

```python
def displot(data, x) -> None:
    sns.displot(data = data, x = x, kde = True, height = 5, color= "royalblue")
    plt.title(f"{x.upper()}", fontdict=title_style)
    plt.grid(axis="y", alpha = 0.7)
    plt.xlabel(x, fontdict=label_style)
    plt.ylabel("frequency", fontdict=label_style)
    plt.axvline(data[x].mean(), color = "orange", ls = "--")
    plt.show()
```

```python
for i in ["age","trtbps","thalach","old peak"]:
    displot(data,i)
```

#### Skewness treatment of old peak feature

```python
data.loc[:,"old peak"] = np.sqrt(data["old peak"])
data
```
